```python
#1. Create a histogram to analyze the distribution of Age.


import matplotlib.pyplot as plt

# Histogram of Age
plt.hist(df['Age'], bins=10)   # you can change bins
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
#2. Plot a pie chart showing the percentage distribution of Gender.



import matplotlib.pyplot as plt

# Count each gender
gender_counts = df['Gender'].value_counts()

# Plot pie chart
plt.pie(gender_counts, 
        labels=gender_counts.index, 
        autopct='%1.1f%%')

plt.title("Gender Distribution")
plt.show()
```

```python
#6. Plot a line chart to visualize Purchase Amount trend over Purchase Date.

import pandas as pd
import matplotlib.pyplot as plt

# Load your file
df = pd.read_csv(r"C:\Users\Dell\Downloads\Practice-Datasets .csv")

# Convert date column to datetime
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# Sort by date (important for trend)
df = df.sort_values('Purchase_Date')

# Line chart
plt.plot(df['Purchase_Date'], df['Purchase_Amount'])

plt.title("Purchase Amount Trend Over Time")
plt.xlabel("Purchase Date")
plt.ylabel("Purchase Amount")

plt.show()

```

```python
#7.7. Create a bar plot to compare average Monthly Income by Gender.

import pandas as pd
import matplotlib.pyplot as plt

# Load file
df = pd.read_csv(r"C:\Users\Dell\Downloads\Practice-Datasets .csv")

# Calculate average income by gender
avg_income = df.groupby('Gender')['Monthly_Income'].mean()

# Bar plot
plt.bar(avg_income.index, avg_income.values)

plt.title("Average Monthly Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Monthly Income")

plt.show()
```

```python
#8. Plot a box plot of Purchase Amount grouped by Product Category.

import matplotlib.pyplot as plt

df.boxplot(column='Purchase_Amount', by='Product_Category')
plt.show()
```

```python
#9. Create a pie chart to show Product Category distribution.

import matplotlib.pyplot as plt

category_counts = df['Product_Category'].value_counts()

plt.pie(category_counts, labels=category_counts.index)
plt.show()
```

```python
#10.10. Plot a histogram of Purchase Amount with appropriate bin size.


import matplotlib.pyplot as plt

plt.hist(df['Purchase_Amount'],
         bins=15,
         edgecolor='black')

plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
#11. Create a bar chart showing average Spending Score by City.
import matplotlib.pyplot as plt

avg_score = df.groupby('City')['Spending_Score'].mean()

plt.bar(avg_score.index, avg_score.values)
plt.show()
```

```python
#12. Plot a line chart showing total Purchase Amount per month.
import matplotlib.pyplot as plt

# Convert to date
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# Total purchase per month
monthly_total = df.groupby(df['Purchase_Date'].dt.to_period('M'))['Purchase_Amount'].sum()

# Line chart
plt.plot(monthly_total.index.astype(str), monthly_total.values)
plt.show()
```

```python
#12. Plot a line chart showing total Purchase Amount per month.
import matplotlib.pyplot as plt

df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

m = df.groupby(df['Purchase_Date'].dt.month)['Purchase_Amount'].sum()

plt.plot(m.index, m.values)
plt.show()
```

```python
#13. Create a box plot comparing Monthly Income across different Cities.

import matplotlib.pyplot as plt

df.boxplot(column='Monthly_Income', by='City')
plt.show()
```

```python
#14. Plot a histogram to analyze Age distribution by Gender.
import matplotlib.pyplot as plt

# Separate ages by gender
m = df[df['Gender'] == 'Male']['Age']
f = df[df['Gender'] == 'Female']['Age']

# Histograms
plt.hist(m, alpha=0.5)
plt.hist(f, alpha=0.5)

plt.show()
```

```python
#15. Create a bar chart showing total Purchase Amount by Product Category.

import matplotlib.pyplot as plt

total = df.groupby('Product_Category')['Purchase_Amount'].sum()

plt.bar(total.index, total.values)
plt.show()
```

```python
#16. Plot a scatter plot between Monthly Income and Spending Score.
import matplotlib.pyplot as plt

plt.scatter(df['Monthly_Income'], df['Spending_Score'])
plt.show()
```

```python
#17. Create a line chart to visualize daily Purchase Amount variation.
import matplotlib.pyplot as plt

df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

daily = df.groupby(df['Purchase_Date'].dt.day)['Purchase_Amount'].sum()

plt.plot(daily.index, daily.values)
plt.show()
```
